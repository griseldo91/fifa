```python
overall_name = []
best_name=''
pos = ''


for pos in df.Position.unique()[:-1]:

    best_overall = df[df.Position == pos].Overall.max()
    best_name = df[(df['Position']== pos) & (df['Overall'] == best_overall) ].Name.iloc[0]
    overall_name.append({'Position':pos,
                        'Name':best_name,
                        'Overall':best_overall})
    
    
    
best_players_df = pd.DataFrame(overall_name) 
```
